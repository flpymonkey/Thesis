```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt

# Specify the path to your CSV file
csv_file_path = 'tensor_graph/ppo_6_rew.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Plot the DataFrame
plt.plot(df['Step'], df['Value']*1e+4)

# Customize the plot
plt.xlabel('Step')
plt.ylabel('Average Reward')
plt.title('Average Episodic Reward')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Specify the path to your CSV file
csv_file_path = 'tensor_graph/actions455_eiie_2019.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

df = df.rename(columns={old_name: i for i, old_name in enumerate(df.columns)})
print(df)


import plotly.graph_objects as go


# Example data (replace with your actual data)
# Ensure data is available and correct
time = df.index

# Calculate cumulative sums for stacking
cash = df[0]
w1 = cash + df[1]
w2 = w1 + df[2]
w3 = w2 + df[3]
w4 = w3 + df[4]
w5 = w4 + df[5]
w6 = w5 + df[6]


# Create the figure
fig = go.Figure()

# Add each trace with fill specified to stack them

# This should stack up to the total load

fig.add_trace(go.Scatter(x=time, y=cash, fill='tozeroy', mode='none', name='Cash'))
fig.add_trace(go.Scatter(x=time, y=w1, fill='tonexty', mode='none', name='AAPL'))
fig.add_trace(go.Scatter(x=time, y=w2, fill='tonexty', mode='none', name='BA'))
fig.add_trace(go.Scatter(x=time, y=w3, fill='tonexty', mode='none', name='INTC'))
fig.add_trace(go.Scatter(x=time, y=w4, fill='tonexty', mode='none', name='MSFT'))
fig.add_trace(go.Scatter(x=time, y=w5, fill='tonexty', mode='none', name='V'))
fig.add_trace(go.Scatter(x=time, y=w6, fill='tonexty', mode='none', name='WMT'))
# fig.add_trace(go.Scatter(x=time, y=load, fill='tonexty', mode='none', name='load'))



# Customize layout
fig.update_layout(
    title='EIIE - Portfolio Weights - OOS 2020-03-13 to 2021-12-30',
    yaxis_title='Weights',
    xaxis_title='Time',
    legend_title='Stock',
    hovermode='x unified'
)

fig.show()
```
