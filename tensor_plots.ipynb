```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'tensor_graph/ppo4feat2020_rew.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Plot the DataFrame
plt.plot(df['Step'], df['Value'])

# Customize the plot
plt.xlabel('Step')
plt.ylabel('Value')
plt.title('Step vs Value')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'tensor_graph/dji_2021_rew.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows of the DataFrame
print(df.head())

# Plot the DataFrame
plt.plot(df['Step'], df['Value']*1e+4)

# Customize the plot
plt.xlabel('Step')
plt.ylabel('Episodic Reward')
plt.title('Average Episodic Reward')
plt.grid(True)

# Show the plot
plt.show()
```
